```python
species = pd.read_csv('species_info.csv')
print(species.dtypes)
print(species.head())
print(species.info())
```

## **Cleaning the DataFrames**

Data cleaning is a crucial step in ensuring the reliability of our analysis. Here, we focus on:
1. **Handling Missing Values**: For instance, the `conservation_status` column contains many `NaN` values, which might require assigning a default status or excluding them from certain analyses.
2. **Removing Duplicates**: Duplicate records can lead to overestimations and skewed insights.
3. **Data Type Consistency**: Ensuring all columns have the appropriate data types for seamless analysis.

### Example
- Before cleaning: The `species_info` DataFrame has approximately 3% of the `conservation_status` column populated.
- After cleaning: We replace `NaN` values with 'No Status' to maintain a consistent dataset.


```python
# sort by the number of observations (smaller first), then keep smallest of duplicates
observations_cleaned = observations.sort_values('observations', ascending=True).drop_duplicates(['scientific_name', 'park_name'], keep= 'first').sort_index() # from https://stackoverflow.com/questions/12497402/remove-duplicates-by-columns-a-keeping-the-row-with-the-highest-value-in-column/68876659#68876659

# Filter the duplicated rows based on 'scientific_name' and sort by 'scientific_name'
duplicated_species = species[species['scientific_name'].duplicated()].sort_values(by='scientific_name')
# Print the duplicated rows
#display(duplicated_species)
duplicate_count = species.duplicated().sum()
# Group by 'scientific_name' and find rows with differing 'common_names' or 'conservation_status'
duplicated_with_differences = duplicated_species.groupby('scientific_name').filter(
    lambda group: len(group) > 1 and (group['common_names'].nunique() > 1 or group['conservation_status'].nunique() > 1)
)
# Print the duplicated rows with differences
#print(duplicated_with_differences)

''' We also see multiple duplicates in the species information. Upon review, we have multiple common names 
and conservation statuses for a single species (ex: Canis lupus is both endangered and in recovery). In a 
similar approch, this analysis will bias towards caution -- marking a species as Endangered rather than 
In Recovery when there is a conflict. All common names have been combined to keep their values intact.'''

# Concatenate common names for each scientific name
species_names = species.groupby(['scientific_name'])['common_names'].apply(','.join).reset_index()

# assign all common names back to species data frame
species_named = pd.merge(species, species_names, on= 'scientific_name', how= 'left')
#print(species_named.head())
# drop old common names column
species_named = species_named.drop(columns= 'common_names_x')

# rename new common names column
species_named.rename(columns= {'common_names_y': 'common_names'}, errors="raise", inplace= True)
# remove redundant names from common names column
species_named['common_names'] = species_named['common_names'].str.replace(' ,', ',')
species_named['common_names'] = species_named['common_names'].str.replace(', ', ',')
species_named['common_names']=species_named['common_names'].str.split(',').apply(lambda x : ', '.join(set(x)))
# encode data to enable sorting by conservation status for duplicates (Endangered > Threatened > Species of Concern > In Recovery > None)
species_named['conservation_status'] = species_named['conservation_status'].fillna("Least Concern")
species_named['conservation_group'] = pd.Categorical(species_named['conservation_status'], ordered=True, categories=['Least Concern', 'In Recovery', 'Species of Concern', 'Threatened', 'Endangered'])
# drop duplicate species keeping the most conservative level in conservation group
species_named = species_named.sort_values('conservation_group', ascending=True).drop_duplicates(['scientific_name'], keep= 'last')
# drop the 'conservation_group' column used for sorting
species_named = species_named.drop('conservation_group',axis=1)
# export the DataFrame to a CSV file
species_named.to_csv('spec_df.csv', index=False)
species_cleaned = species_named
```

```python
## Merging both dataframes
biodiversity_data = pd.merge(species_cleaned, observations_cleaned, on= 'scientific_name', how= 'left')
#display(biodiversity_data.info())
#print(biodiversity_data['park_name'].unique())
# Rename reductant information on park_names
park_renames = {
    'Bryce National Park': 'Bryce',
    'Yellowstone National Park': 'Yellowstone',
    'Great Smoky Mountains National Park': 'Great Smoky Mountains',
    'Yosemite National Park': 'Yosemite'
}

biodiversity_data['park_name'] = biodiversity_data['park_name'].replace(park_renames)
biodiversity_data.to_csv('biodiversity_data.csv', index=False)
#display(biodiversity_data.head())
#print(biodiversity_data[biodiversity_data['scientific_name'] == 'Canis lupus'])
biodiversity_data.describe(include='all')
```

### **Conservation Statuses and Categories**

Conservation statuses provide critical information about the health of species populations. The main categories analyzed in this project are:

- **Least Concern**: Species with a stable population and low risk of extinction.
- **Species of Concern**: Species requiring monitoring but not at immediate risk.
- **Endangered**: Species facing a high risk of extinction in the near future.
- **Threatened**: Species likely to become endangered.
- **In Recovery**: Species whose populations are improving due to conservation efforts.

### **Analysis Goals**
This project aims to:
1. Understand the distribution of species across these categories.
2. Identify which categories are most prevalent in national parks.
3. Visualize patterns across parks and species groups.

### **Visualization**

Visualizations are used to effectively communicate insights from the data. The following plots are included in the analysis:
- **Bar Charts**:
  - Illustrate the count of species in each conservation status category.
  - Compare the number of species across different categories within each conservation status.
- **Box Plots**:
  - Explore the distribution of species observations across parks for each status.
- **Histograms**:
  - Show the frequency distribution of species counts within categories to identify trends.

These visualizations provide an in-depth view of the species' conservation statuses and their distributions across parks.

### **Insights**

From the visualization of observations by conservation status:
- **Least Concern**: Dominates the dataset, reflecting the majority of species in stable conditions.
- **Endangered and Threatened**: Represent a small fraction of species, indicating the importance of focused conservation efforts.
- **Species of Concern**: Highlights a notable group that requires monitoring but is not yet at immediate risk.


```python
# Define categories with some degree of endangerment
statuses = list(biodiversity_data.conservation_status.unique())
print(statuses)
endangered_statuses = ['Species of Concern', 'Endangered', 'Threatened', 'In Recovery']
no_concern = biodiversity_data[biodiversity_data['conservation_status'] == 'Least Concern']
# Filter the DataFrame to include only the rows with the specified categories
endangerment = biodiversity_data[biodiversity_data['conservation_status'].isin(endangered_statuses)]
species_of_concern = endangerment[endangerment['conservation_status'] == 'Species of Concern'];
endangered = endangerment[endangerment['conservation_status'] == 'Endangered'];
threatened = endangerment[endangerment['conservation_status'] == 'Threatened'];
in_recovery = endangerment[endangerment['conservation_status'] == 'In Recovery'];
# Display the new DataFrame
#print(endangered.head())
print(no_concern.head())
```

```python
# Determine grid size
n_statuses = len(endangered_statuses)
n_cols = 2
n_rows = (n_statuses + n_cols - 1) // n_cols  # Calculate the number of rows dynamically

# Create a figure for the endangered statuses
fig = plt.figure(figsize=(15, 10))

# First loop: Plot for statuses other than 'Least Concern'
for i, status in enumerate(endangered_statuses, start=1):
    # Filter the DataFrame for the current status
    subset = endangerment[endangerment['conservation_status'] == status]
    
    # Create a subplot
    ax = plt.subplot(n_rows, n_cols, i)  # Adjust grid dynamically
    
    # Generate the bar plot
    sns.countplot(x='category', data=subset, hue='category', palette='Spectral', ax=ax)
    
    # Configure title and labels
    ax.set_title(f'{status} By Category')
    ax.set_xlabel('Category')
    ax.set_ylabel('Number')
    plt.xticks(rotation=30, fontsize=8)

# Adjust layout and save the plot before showing it
plt.tight_layout()
plt.savefig('Endangered Species.png', dpi=600)
plt.show()  # Show after saving

# Second plot: Only for 'Least Concern'
plt.figure(figsize=(15, 5))
subset = no_concern[no_concern['conservation_status'] == 'Least Concern']

# Single bar plot for 'Least Concern'
sns.countplot(x='category', data=subset, hue='category', palette='Spectral')
plt.title('Least Concern By Category')
plt.xlabel('Category')
plt.ylabel('Number')
plt.xticks(rotation=30, fontsize=8)

# Save and show the second plot
plt.tight_layout()
plt.savefig('Least Concern Species.png', dpi=600)
plt.show()
```

```python
# Create a histplots of species observations by category by park

bio_df_short = biodiversity_data.replace({'Bryce National Park': 'Bryce',
                               'Great Smoky Mountains National Park': 'G.S.M.',
                               'Yosemite National Park': 'Yosemite',
                               'Yellowstone National Park': 'Yellowstone'})

categories = ['Mammal', 'Vascular Plant', 'Bird', 'Amphibian', 'Reptile', 'Nonvascular Plant', 'Fish']
parks = ['Great Smoky Mountains', 'Bryce', 'Yosemite', 'Yellowstone']

fig = plt.figure(figsize=(10, 12))
plt.subplots_adjust(hspace= 0.40, wspace= 0.40)
plt.suptitle('Observation Distributions by Category & Park', fontsize=18, y=1)
fig.text(0.5, 0.005, 'Observations', ha="center", va='center', fontsize=14)

i = 0
# loop through the length of statuses and keep track of indices
for m, category in enumerate(categories, 1):
    for n, park in enumerate(parks, 1):
           
        i = i + 1
        ax = plt.subplot(7, 4, i)
        
        # add a new subplot iteratively
        plot_data = bio_df_short[(bio_df_short.category == category) & (bio_df_short.park_name == park)]
        sns.histplot(x= 'observations', data=plot_data, color='skyblue', edgecolor='white')

        # formatting the figures & plots
        ax.set_title(park.title())
        ax.set_xlabel('')
        ax.set_ylabel(category.title())
                
fig.tight_layout()
plt.savefig('Observation Distributions by Category & Park.png', dpi= 600)
```

```python
# plot boxplots of normalized observations by status for each park
bio_df_troubled = biodiversity_data.loc[biodiversity_data['conservation_status'] != 'Least Concern']
normed_df = bio_df_troubled.copy()

parks = ['Great Smoky Mountains', 'Bryce', 'Yosemite', 'Yellowstone']

for park in parks:
    park_obs_min = normed_df['observations'][normed_df['park_name'] == park].min()
    park_obs_max = normed_df['observations'][normed_df['park_name'] == park].max()
    park_obs_rng = park_obs_max - park_obs_min
    for idx, row in normed_df.iterrows():
        if  normed_df.loc[idx,'park_name'] == park:
            normed_df.loc[idx, 'obs2'] = (normed_df.loc[idx,'observations'] - park_obs_min) / park_obs_rng
            
```

```python
fig = plt.figure(figsize=(12, 7))
plt.subplots_adjust(hspace= 0.25, wspace= 0.30)
plt.title('Normalized Distribution of Status\' Observations by Park', fontsize=18, y=1.025)

# filter df and plot each status's observations by plot on the new subplot axes
ax = sns.boxplot(data= normed_df, 
                 x= 'obs2', 
                 y= 'park_name', 
                 hue= 'conservation_status', 
                 width= 0.8, 
                 palette= 'Spectral_r', 
                 order= ['Great Smoky Mountains', 'Bryce', 'Yosemite', 'Yellowstone'])

# formatting the figures & plots
ax.set_ylabel('')
ax.set_xlabel('')
plt.xlabel('Observations', fontsize=14)
plt.xticks(fontsize= 12)
plt.yticks(fontsize= 12)
plt.yticks(rotation = 35)
plt.legend(title= 'Conservation Status', bbox_to_anchor=(1.03,1), loc="upper left")

plt.savefig('Normalized Distribution of Status Observations by Park', dpi= 600, bbox_inches='tight')
```
